Replacing all the NaN values with 0 (0, 0) for easier analysis

```python
Social_data['Users_United_States'].fillna("0 (0000, 00)", inplace = True)
Social_data['Users_Indonesia'].fillna("0 (0000, 00)", inplace = True)
Social_data['Users_Singapore'].fillna("0 (0000, 00)", inplace = True)
Social_data['Users_India'].fillna("0 (0000, 00)", inplace = True)
Social_data['Users_Vietnam'].fillna("0 (0000, 00)", inplace = True)
Social_data['Users_Philippines'].fillna("0 (0000, 00)", inplace = True)
Social_data['Users_Bangladesh'].fillna("0 (0000, 00)", inplace = True)
```

```python
Social_data.iloc[8,6] = "600000 (0000, 00)"
Social_data.iloc[9,6] = "6200000 (2020, 71)"
```

Now we need to split every column with Number of Users and Survey Year of respective country for better understanding and good data type representation.

```python
def UserNumber(x):
  y= str(x)
  z=y.split(' (')
  z=z[0]
  return (z[:])
Social_data['Users_in_US']=Social_data['Users_United_States'].apply(UserNumber)
Social_data['Users_in_Indonesia']=Social_data['Users_Indonesia'].apply(UserNumber)
Social_data['Users_in_Singapore']=Social_data['Users_Singapore'].apply(UserNumber)
Social_data['Users_in_China']=Social_data['Users_china'].apply(UserNumber)
Social_data['Users_in_India']=Social_data['Users_India'].apply(UserNumber)
Social_data['Users_in_Vietnam']=Social_data['Users_Vietnam'].apply(UserNumber)
Social_data['Users_in_Philippines']=Social_data['Users_Philippines'].apply(UserNumber)
Social_data['Users_in_Bangladesh']=Social_data['Users_Bangladesh'].apply(UserNumber)
```

```python
Social_data.head(10)
```

```python
def SurveyYearandRank(x):
  y= str(x)
  z=y.split(' (')
  z=z[1]
  return (z[:])
Social_data['SurveyYear_Rank_US']=Social_data['Users_United_States'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_Indonesia']=Social_data['Users_Indonesia'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_Singapore']=Social_data['Users_Singapore'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_China']=Social_data['Users_china'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_India']=Social_data['Users_India'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_Vietnam']=Social_data['Users_Vietnam'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_Philippines']=Social_data['Users_Philippines'].apply(SurveyYearandRank)
Social_data['SurveyYear_Rank_Bangladesh']=Social_data['Users_Bangladesh'].apply(SurveyYearandRank)
```

```python
Social_data.head(10)
```

### Now we can drop the initial columns and reduce the observation number :

```python
Social_data.drop('Users_United_States', axis=1, inplace=True)
Social_data.drop('Users_Indonesia', axis=1, inplace=True)
Social_data.drop('Users_Singapore', axis=1, inplace=True)
Social_data.drop('Users_china', axis=1, inplace=True)
Social_data.drop('Users_India', axis=1, inplace=True)
Social_data.drop('Users_Vietnam', axis=1, inplace=True)
Social_data.drop('Users_Philippines', axis=1, inplace=True)
Social_data.drop('Users_Bangladesh', axis=1, inplace=True)
```

```python
Social_data.head(10)
```

### To further refine the data set we will now split the SurveyYear_Rank columns to SurveyYear and Rank for further analysis -

```python
def SurveyYear(x):
  y= str(x)
  z=y.split(', ')
  z=z[0]
  return (z[:])
Social_data['SurveyYear_US']=Social_data['SurveyYear_Rank_US'].apply(SurveyYear)
Social_data['SurveyYear_Indonesia']=Social_data['SurveyYear_Rank_Indonesia'].apply(SurveyYear)
Social_data['SurveyYear_Singapore']=Social_data['SurveyYear_Rank_Singapore'].apply(SurveyYear)
Social_data['SurveyYear_China']=Social_data['SurveyYear_Rank_China'].apply(SurveyYear)
Social_data['SurveyYear_India']=Social_data['SurveyYear_Rank_India'].apply(SurveyYear)
Social_data['SurveyYear_Vietnam']=Social_data['SurveyYear_Rank_Vietnam'].apply(SurveyYear)
Social_data['SurveyYear_Philippines']=Social_data['SurveyYear_Rank_Philippines'].apply(SurveyYear)
Social_data['SurveyYear_Bangladesh']=Social_data['SurveyYear_Rank_Bangladesh'].apply(SurveyYear)
```

```python
Social_data.head(10)
```

```python
def Rank(x):
  y= str(x)
  z=y.split(', ')
  z=z[1]
  return (z[:2])
Social_data['Rank_US']=Social_data['SurveyYear_Rank_US'].apply(Rank)
Social_data['Rank_Indonesia']=Social_data['SurveyYear_Rank_Indonesia'].apply(Rank)
Social_data['Rank_Singapore']=Social_data['SurveyYear_Rank_Singapore'].apply(Rank)
Social_data['Rank_China']=Social_data['SurveyYear_Rank_China'].apply(Rank)
Social_data['Rank_India']=Social_data['SurveyYear_Rank_India'].apply(Rank)
Social_data['Rank_Vietnam']=Social_data['SurveyYear_Rank_Vietnam'].apply(Rank)
Social_data['Rank_Philippines']=Social_data['SurveyYear_Rank_Philippines'].apply(Rank)
Social_data['Rank_Bangladesh']=Social_data['SurveyYear_Rank_Bangladesh'].apply(Rank)
```

```python
Social_data.head(10)
```

### Now it looks more refined for Analysis so we can drop the SurveyYear_Rank columns for eacg respective country as we have separated the data as per our requirement -

```python
Social_data.drop('SurveyYear_Rank_US', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_Indonesia', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_Singapore', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_China', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_India', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_Vietnam', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_Philippines', axis=1, inplace=True)
Social_data.drop('SurveyYear_Rank_Bangladesh', axis=1, inplace=True)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.worldometers.info/world-population/us-population/'
requests.get(url)
page = requests.get(url)

soup = BeautifulSoup(page.text, 'lxml')
print(soup)

table_data = soup.find('table', class_ = 'table table-striped table-bordered table-hover table-condensed table-list')

headers = []
for i in table_data.find_all('th'):
    title = i.text
    headers.append(title)

us_population = pd.DataFrame(columns = headers)

for j in table_data.find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [tr.text for tr in row_data]
        length = len(us_population)
        us_population.loc[length] = row
```
